## Plotting Routines

### 1D Data

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

# Create figure with 4 rows, 2 columns
fig, axes = plt.subplots(4, 2, figsize=(10, 12))

# Scatter plot (Top-left)
axes[0,0].scatter(x, y, color='red', label="Scatter")
axes[0,0].set_title("Scatter Plot")
axes[0,0].legend()

# Bar plot (Top-right)
axes[0,1].bar([1,2,3], [3,4,5], color='blue', label="Bar")
axes[0,1].set_title("Bar Chart")
axes[0,1].legend()

# Horizontal bar plot (Middle-left)
axes[1,0].barh([0.5,1,2.5], [0,1,2], color='green', label="BarH")
axes[1,0].set_title("Horizontal Bar Chart")
axes[1,0].legend()

# Horizontal line (Middle-right)
axes[1,1].axhline(0.45, color='purple', linestyle="--", label="AxHLine")
axes[1,1].set_title("Horizontal Line")
axes[1,1].legend()

# Vertical line (Bottom-left)
axes[2,0].axvline(0.65, color='orange', linestyle="--", label="AxVLine")
axes[2,0].set_title("Vertical Line")
axes[2,0].legend()

# Fill between (Bottom-right)
axes[2,1].fill_between(x, y, color='yellow', alpha=0.5, label="FillBetween")
axes[2,1].set_title("Fill Between")
axes[2,1].legend()

# Line plot (New subplot in row 4, col 1)
axes[3,0].plot(x, y, color='black', label="Line Plot")
axes[3,0].set_title("Line Plot")
axes[3,0].legend()

# Fill plot (New subplot in row 4, col 2)
axes[3,1].fill(x, y, color='blue', alpha=0.5, label="Fill")
axes[3,1].set_title("Fill Plot")
axes[3,1].legend()

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

### 2D Data

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
data = np.random.rand(10, 10) * 2
data2 = np.random.rand(10, 10) * 3
Y, X = np.mgrid[-3:3:100j, -3:3:100j]
U = -1 - X**2 + Y
V = 1 + X - Y**2

# Create figure with 3 rows, 2 columns
fig, axes = plt.subplots(3, 2, figsize=(12, 14))

# 1. Image display (Top-left)
img = np.random.rand(10, 10)  # Placeholder image
axes[0,0].imshow(img, cmap='gist_earth', interpolation='nearest', vmin=-2, vmax=2)
axes[0,0].set_title("Image Display")

# 2. Pseudo-color plot (Top-right)
axes[0,1].pcolor(data2, cmap='coolwarm')
axes[0,1].set_title("Pcolor Plot")

# 3. Pseudo-color mesh (Middle-left)
axes[1,0].pcolormesh(data, cmap='viridis')
axes[1,0].set_title("Pcolormesh Plot")

# 4. Contour plot (Middle-right)
CS = axes[1,1].contour(Y, X, U, cmap='plasma')
axes[1,1].set_title("Contour Plot")
axes[1,1].clabel(CS, inline=True, fontsize=8)  # Label contours

# 5. Filled contour plot (Bottom-left)
axes[2,0].contourf(data, cmap='inferno')
axes[2,0].set_title("Filled Contour Plot")


plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

### Vector Fields

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
X, Y = np.meshgrid(np.linspace(-2, 2, 10), np.linspace(-2, 2, 10))
U = -1 - X**2 + Y
V = 1 + X - Y**2

fig, axes = plt.subplots(3, 1, figsize=(8, 12))

# 1. Arrow plot
axes[0].arrow(0, 0, 0.5, 0.5, head_width=0.1, head_length=0.1, fc='red', ec='black')
axes[0].set_xlim(-1, 1)
axes[0].set_ylim(-1, 1)
axes[0].set_title("Arrow Plot")

# 2. Quiver plot
axes[1].quiver(X, Y, U, V, color='blue')
axes[1].set_title("Quiver Plot")

# 3. Stream plot
axes[2].streamplot(X, Y, U, V, color='green')
axes[2].set_title("Stream Plot")

plt.tight_layout()
plt.show()
```

### Data Distributions

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
y = np.random.normal(0, 1, 100)  # Normally distributed data
z = np.random.normal(0, 1, 100)  # Another dataset for violin plot

# Create figure with 3 rows, 1 column
fig, axes = plt.subplots(3, 1, figsize=(8, 12))

# 1. Histogram
axes[0].hist(y, bins=20, color='blue', alpha=0.7, edgecolor='black')
axes[0].set_title("Histogram")

# 2. Box Plot
axes[1].boxplot(y, vert=True, patch_artist=True)
axes[1].set_title("Box Plot")

# 3. Violin Plot
axes[2].violinplot(z, showmeans=True, showmedians=True)
axes[2].set_title("Violin Plot")

plt.tight_layout()
plt.show()
```

## Customization Plot

### Color, Color Bars and Color Maps

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
img = np.random.rand(10, 10)  # Placeholder for image data

# Create figure with 3 rows, 2 columns
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# 1. Multiple Line Plot (Top-left)
axes[0,0].plot(x, x, label="y=x")
axes[0,0].plot(x, x**2, label="y=x²")
axes[0,0].plot(x, x**3, label="y=x³")
axes[0,0].set_title("Line Plot: x, x², x³")
axes[0,0].legend()

# 2. Alpha Transparency Line Plot (Top-right)
axes[0,1].plot(x, y, alpha=0.4, label="Alpha=0.4", color='blue')
axes[0,1].set_title("Transparent Line Plot")
axes[0,1].legend()

# 3. Black Line Plot (Middle-left)
axes[1,0].plot(x, y, color='k', label="Black Line")
axes[1,0].set_title("Black Line Plot")
axes[1,0].legend()

# 4. Image Display with Seismic Colormap (Middle-right)
im = axes[1,1].imshow(img, cmap='seismic')
fig.colorbar(im, ax=axes[1,1], orientation='horizontal')
axes[1,1].set_title("Image with Seismic Colormap")


plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

### Markers

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

# Create figure with 2 rows, 1 column
fig, axes = plt.subplots(2, 1, figsize=(8, 10))

# 1. Scatter plot with dot markers
axes[0].scatter(x, y, marker='.', color='red')
axes[0].set_title("Scatter Plot (Marker='.')")

# 2. Line plot with circle markers
axes[1].plot(x, y, marker='o', color='blue')
axes[1].set_title("Line Plot (Marker='o')")

plt.tight_layout()
plt.show()
```

### Linestyles

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

# Create figure with 3 rows, 2 columns
fig, axes = plt.subplots(3, 2, figsize=(12, 10))

# 1. Thick Line Plot
axes[0,0].plot(x, y, linewidth=4.0, color='blue')
axes[0,0].set_title("Thick Line (linewidth=4.0)")

# 2. Solid Line Plot
axes[0,1].plot(x, y, ls='solid', color='green')
axes[0,1].set_title("Solid Line (ls='solid')")

# 3. Dashed Line Plot
axes[1,0].plot(x, y, ls='--', color='purple')
axes[1,0].set_title("Dashed Line (ls='--')")

# 4. Multiple Line Styles
axes[1,1].plot(x, y, '--', label="Dashed")
axes[1,1].plot(x**2, y**2, '-.', label="Dash-dot")
axes[1,1].set_title("Mixed Line Styles")
axes[1,1].legend()

# 5. Step Plot
axes[2,0].step(x, y, color='red', linewidth=4.0)
axes[2,0].set_title("Step Plot")

plt.tight_layout()
plt.show()
```

### Text and Annotations

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

# Create figure and axis
fig, ax = plt.subplots(figsize=(8, 6))

# Plot sine wave
ax.plot(x, y, color='blue', label="Sine Wave")

# Add text annotation
ax.text(1, -2.1, 'Example Graph', style='italic', fontsize=12)

# Add arrow annotation
ax.annotate('Sine', 
            xy=(np.pi, 0),  # Approximate x-value for zero crossing
            xycoords='data',
            xytext=(4, 0.5),  # Adjust text position
            textcoords='data',
            arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='red'))

# Customize plot
ax.set_title("Annotated Sine Wave")
ax.set_xlabel("X values")
ax.set_ylabel("Y values")
ax.legend()

plt.show()
```

### Limits, Legends and Layouts

#### Limits and autoscaling

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create figure and axis
fig, ax = plt.subplots(figsize=(8, 6))

# Plot sine wave
ax.plot(x, y, color='blue', label="Sine Wave")

# Adjust margins
ax.margins(x=0.0, y=0.1)  # No extra margin on x, 10% margin on y

# Set equal aspect ratio
ax.axis('equal')

# Set axis limits
ax.set(xlim=[0, 10.5], ylim=[-1.5, 1.5])  # Alternative: ax.set_xlim(0, 10.5)

# Customize plot
ax.set_title("Axis Customization in Matplotlib")
ax.set_xlabel("X values")
ax.set_ylabel("Y values")
ax.legend()

plt.show()
```

#### Legends

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create figure and axis
fig, ax = plt.subplots(figsize=(8, 6))

# Plot two lines with labels for legend
ax.plot(x, y1, label="Sine Wave", color='blue')
ax.plot(x, y2, label="Cosine Wave", color='red')

# Apply title, labels, and legend
ax.set(title='An Example Axes', xlabel='X-Axis', ylabel='Y-Axis')
ax.legend(loc='best')  # Automatically finds best location

plt.show()
```

#### Ticks

```python
import matplotlib.pyplot as plt

# Create figure and axis
fig, ax = plt.subplots()

# Plot some example data
ax.plot(range(1, 5), [10, 20, 30, 40], marker='o')

# Customize x-axis ticks and labels
ax.xaxis.set(ticks=range(1, 5), ticklabels=[3, 100, -12, 'foo'])

# Customize y-axis tick appearance
ax.tick_params(axis='y', direction='inout', length=10)

# Show the plot
plt.show()
```

### Subplot spacing

```python
import matplotlib.pyplot as plt

# Create figure and subplots
fig3, ax = plt.subplots(1, 1, figsize=(6, 4))

# Customize x-axis ticks and labels
ax.xaxis.set(ticks=range(1, 5), ticklabels=[3, 100, -12, 'foo'])

# Customize y-axis tick parameters
ax.tick_params(axis='y', direction='inout', length=10)

# Adjust subplot layout manually
fig3.subplots_adjust(wspace=0.5,
                     hspace=0.3,
                     left=0.125,
                     right=0.9,
                     top=0.9,
                     bottom=0.1)

# Optimize layout
fig3.tight_layout()

# Show the plot
plt.show()
```

### Axis Spines

```python
import matplotlib.pyplot as plt

# Create figure and axis
fig, ax1 = plt.subplots(figsize=(6, 4))

# Customize x-axis ticks and labels
ax1.xaxis.set(ticks=range(1, 5), ticklabels=[3, 100, -12, 'foo'])

# Customize y-axis tick parameters
ax1.tick_params(axis='y', direction='inout', length=10)

# Adjust spine visibility and position
ax1.spines['top'].set_visible(False)  # Hide the top spine
ax1.spines['bottom'].set_position(('outward', 10))  # Shift bottom spine outward

# Adjust subplot layout
fig.subplots_adjust(wspace=0.5, hspace=0.3, left=0.125, right=0.9, top=0.9, bottom=0.1)

# Optimize layout
fig.tight_layout()

# Show the plot
plt.show()
```
